### input two integger and tell the relationship beetwin numbers != , ==, <,>

```python
'''input two integer to tell the relation beetwin numbers '''

first_number = int(input ('Enter first number: '))
second_number= int(input('Enter second number: '))

if first_number == second_number:
    print(f"the  {first_number} and {second_number} are same")
elif first_number > second_number:
    print(f"the  {first_number}  is bigger and not equal to  {second_number} ")
elif first_number < second_number:
    print(f"the  {first_number}  is smaller and not equal to  {second_number} ")

```
