## Question 4

```python
# loop
while 1:
    # Error Handeling
    try:
        #input
        grade = float(input("Enter student's grade: "))
        
        # just in range 0 to 100
        if 100 >= grade >=0 :

            # conditions to choose Level
            #for less than 70 = D
            if 70 > grade :
                score = "D"

            # for 79 to 70 = C
            elif 79>= grade >=70:
                score = "C"

            #for 89 to 80 = B
            elif 89 >= grade >= 80:
                score = "B"
            
            # for more than 89 = A
            else:
                score = "A"
            #output
            print(score)
            # after the job completed the loop will break
            break
        else:
            # raise Error if grade be out of range
            raise
    except:
        # alert message
        print('Only numeric values for grade in range 0 to 100 ')
```

## Question 6

```python
# loop
while 1:
    #Error handeling
    try:
        # loop break
        loop_break = input("\"S\" for stop or Enter to continue").lower()
        if loop_break =="s":
            break

        #user inputs variables
        num1 = float(input('Enter a number to calculate: '))
        num2 = float(input('Enter another number: '))
        operators = input('+  -  *  / :(enter that operator you want ) ')
        # conditions with match cases
        match operators:
            case "+":
                # output
                print(f"{num1} + {num2} = {num1+num2}")
            case "-":
                # output
                print(f"{num1} - {num2} = {num1-num2}")
            case "*":
                # output
                print(f"{num1} * {num2} = {num1*num2}")
            case "/":
                # handel divid by zero error 
                try:
                    # output
                    print(f"{num1} ÷ {num2} = {num1/num2}")
                except ZeroDivisionError:
                    # Error message
                    print("You can't divid any thing by \"Zero\"")
    except:
        #Error message
        print('Only Numeric Valuse')
```
