## Question 11

```python
# variable
phone_book = {}

# loop 
while 1:
    flag = input('1. Add 2.search 3.show 4.exit')
    if flag == "4":
        # exit
        break
    elif flag == "1":
        #add
        name = input('Enter name: ')
        phone_number = input('Enter phone number: ')
        phone_book[name] = phone_number
    elif flag == "2":
        # search
        name = input('Enter name: ')
        print(phone_book.get(name.lower(),f'{name} is not in contact list'))

    elif flag == "3":
        #show
        for people,number in zip(phone_book.keys(),phone_book.values()):
            print(f"{people} -- {number}",end='\n')
```

## Question 19

```python
from functools import reduce

def flatten_reduce_lambda(matrix):
    return list(reduce(lambda x, y: x + y, matrix, []))

with open(r"E:\Maktab-145\Datasets\q18.txt") as f:
    lines = f.read()
    users = flatten_reduce_lambda([line.split(" - ") for line in lines.split("\n")])
age =[]
for user in users:
    age.append(int(user.split('-')[1]))

average_age = sum(age)/len(age)
print(age)
print(average_age)
```

## Question 20

```python
## Question 20
student_info = {}

def add():
    global student_info
    try:
        name = input("Please enter your name: ")
        student_info[name] = float(input("Please enter your grade: "))
    except ValueError:
        print("Please enter a numeric value")
        add()

def save():
    with open(r"E:\Maktab-145\Datasets\student_info.txt", "a+") as file:
        for name in student_info:
            file.write(f"{name}: {student_info[name]}\n")
    student_info.clear()

def show():
    with open(r"E:\Maktab-145\Datasets\student_info.txt", "r") as file:
        print(file.read())

def search(text):
    with open(r"E:\Maktab-145\Datasets\student_info.txt", "r") as file:
        lines = file.readlines()
        
        for line in lines:
            if text.lower() in line.lower():
                print(line)

while True:
    command = input("Please enter your command (add, save, show, search, exit): ").lower()
    match command:
        case "add":
            add()
        case "save":
            save()
        case "show":
            show()
        case "search":
            search_text = input("Please enter your search query: ")
            search(search_text)
        case "exit":
            break
```
