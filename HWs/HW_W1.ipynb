# Question 4  `random number game`

```python
import random

health = 5
system_guess = random.randint(1,20)

while health!=0:
    print(f"Your health is {health}❤️")
    user_guess = int(input('guess a number: '))

    if user_guess < system_guess:
        print('guess bigger!')
        
    elif user_guess > system_guess:
        print('guess smaller')
    elif user_guess == system_guess:
        print('You won!!!')
        break
    health-=1
```

# Question 5 `library management system`

```python
#imports
import json
# functions 

#remove book
def remove(book_name):
    if book_name in library:
        del library[book_name]
        save()
        print("Book removed successfully")
    else:
        print("Book not found")
#read json file with error handeling
def json_reader(path):
    try:
        with open(path, "r", encoding="utf-8") as file:
            library = json.load(file)
            return library
    except FileNotFoundError:
        print('The file does not exist')
        return {}

    except json.JSONDecodeError:
        print('Invalid JSON file')
        return {}

# add function 
def add (book_name,author):
    library[book_name]=author
# search function
def search (book_name):
    print(library.get(book_name,f"{book_name} not found"))
#show function
def show():
    for i,book in enumerate(library.items(),start=1):
        print(f"{i}.Book name: {book[0]},Author {book[1]}")
#save function
def save():
    with open(r"E:\Maktab-145\Datasets\library.json",'w')as file:
        json.dump(library,file,ensure_ascii=False,indent=4)

library = json_reader(path=r"E:\Maktab-145\Datasets\library.json")

# loop user interface
while True:
    # options
    option = input("1.add 2.search 3.show 4.save 5.remove 6.exit")
    match option:
        #add case
        case '1':
            book_name = input("Enter book name: ")
            author = input("author name: ")
            add(book_name=book_name,author=author)
        #search case
        case '2':
            book_name = input('Enter book name: ')
            search(book_name=book_name)
        #show case
        case '3':
            show()
        #save case
        case '4':
            save()
        #remove a book
        case '5':
            book_name = input('Enter book name: ')
            remove(book_name=book_name)
        #exit case
        case '6':
            save()
            break

    
```

# Question 6 `Store Inventory Management System` 

```python
#functions 
#========================================================================
#add
def add (product_name,quantity):
    if product_name not in products_dictionary:
        products_dictionary[product_name] = quantity
    else:
        products_dictionary[product_name] += quantity

#save
def save():
    with open(r'E:\Maktab-145\Datasets\store.txt','w') as file:
        for line in products_dictionary.items():
            file.write(f"{line[0]} - {line[1]}\n")
    

#show
def show():
    for i,(product,quantity)  in enumerate(products_dictionary.items(),start=1):
        print(f"{i}. {product} - {quantity}\n🟰🟰🟰🟰🟰🟰")
    print("\n")
#report
def report():
    #variables
    max_available =  max(products_dictionary, key=products_dictionary.get)
    min_available =  min(products_dictionary, key=products_dictionary.get)
    # show sum of available products
    print(f"Sum of all kind products = {len (products_dictionary)}\n🟰🟰🟰🟰🟰🟰\n\
Sum of all available products = {sum(products_dictionary.values())}\n🟰🟰🟰🟰🟰🟰\n\
the product with most quantity  {max_available} = {products_dictionary[max_available]}\n🟰🟰🟰🟰🟰🟰\n\
the product with most quantity  { min_available} = {products_dictionary[min_available]}\n🟰🟰🟰🟰🟰🟰\n")
    
#sell
def sell(product_name,quantity):
    if product_name in products_dictionary:
        if products_dictionary[product_name] - quantity >= 0:
            products_dictionary[product_name] -= quantity
            if products_dictionary[product_name] == 0:
                del products_dictionary[product_name]
        else:
            print('Not enough stock available')
    else:
        print('This product is not in list')
#search
def search(product_name):
    if product_name in products_dictionary:
        print(f"{product_name} - {products_dictionary[product_name]}")
    else:
        print(f'{product_name} Not found')
#read file
def read_file():
    products_dictionary = {}

    try:
        with open(r'E:\Maktab-145\Datasets\store.txt', 'r') as file:

            for line in file:

                line = line.strip()

                if not line:
                    continue

                product_name, quantity = line.split('-')

                products_dictionary[product_name.strip()] = int(quantity.strip())

    except FileNotFoundError:
        print("Store file not found. Creating a new inventory.")

    return products_dictionary
        

#========================================================================
#products dictionary
products_dictionary = read_file()

#========================================================================
# loop user interaface
while True:
    #options
    option = input('1.add 2.sell 3.show 4.search 5.save 6.report 7.exit: ')
    #match cases
    match option:
        # add
        case '1':
            try:
                product_name_input = input("Enter product name: ")
                product_quantity_input = int(input("Enter quantity: "))
                add (product_name=product_name_input,quantity=product_quantity_input)
            except ValueError:
                print('You should enter numeric values for quantity')
        # sell
        case '2':
            try:
                product_name_input = input("Enter product name: ")
                product_quantity_input = int(input("Enter quantity: "))
                sell (product_name=product_name_input,quantity=product_quantity_input)
            except ValueError:
                print('You should enter numeric values for quantity')

        # show
        case '3':

            show()

        # search
        case '4':

            product_name_input = input("Enter product name: ")
            search (product_name=product_name_input)
        # save
        case '5':

            save()

        # report
        case '6':

            report()

        # exit
        case '7':
            save()
            break
```
