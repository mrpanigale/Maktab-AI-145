# Question 4  `random number game`

```python
import random

health = 5
system_guess = random.randint(1,20)

while health!=0:
    print(f"Your health is {health}❤️")
    user_guess = int(input('guess a number: '))

    if user_guess < system_guess:
        print('guess bigger!')
        
    elif user_guess > system_guess:
        print('guess smaller')
    elif user_guess == system_guess:
        print('You won!!!')
        break
    health-=1
```

# Question 5 `library management system`

```python
#imports
import json
# functions 

#remove book
def remove(book_name):
    if book_name in library:
        del library[book_name]
        save()
        print("Book removed successfully")
    else:
        print("Book not found")
#read json file with error handeling
def json_reader(path):
    try:
        with open(path, "r", encoding="utf-8") as file:
            library = json.load(file)
            return library
    except FileNotFoundError:
        print('The file does not exist')
        return {}

    except json.JSONDecodeError:
        print('Invalid JSON file')
        return {}

# add function 
def add (book_name,author):
    library[book_name]=author
# search function
def search (book_name):
    print(library.get(book_name,f"{book_name} not found"))
#show function
def show():
    for i,book in enumerate(library.items(),start=1):
        print(f"{i}.Book name: {book[0]},Author {book[1]}")
#save function
def save():
    with open(r"E:\Maktab-145\Datasets\library.json",'w')as file:
        json.dump(library,file,ensure_ascii=False,indent=4)

library = json_reader(path=r"E:\Maktab-145\Datasets\library.json")

# loop interface
while True:
    # options
    option = input("1.add 2.search 3.show 4.save 5.remove 6.exit")
    match option:
        #add case
        case '1':
            book_name = input("Enter book name: ")
            author = input("author name: ")
            add(book_name=book_name,author=author)
        #search case
        case '2':
            book_name = input('Enter book name: ')
            search(book_name=book_name)
        #show case
        case '3':
            show()
        #save case
        case '4':
            save()
        #remove a book
        case '5':
            book_name = input('Enter book name: ')
            remove(book_name=book_name)
        #exit case
        case '6':
            save()
            break

    
```
